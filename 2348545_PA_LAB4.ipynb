```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import Pipeline

# 1. Preprocessing
# Dropping irrelevant columns
data_clean = data.drop(columns=['ID', 'Name', 'Club', 'Contract Valid Until', 'Joined'])

# Label encoding for categorical columns
le_position = LabelEncoder()
data_clean['Position'] = le_position.fit_transform(data_clean['Position'])

le_foot = LabelEncoder()
data_clean['Preferred Foot'] = le_foot.fit_transform(data_clean['Preferred Foot'])

le_nationality = LabelEncoder()  # Encoding Nationality
data_clean['Nationality'] = le_nationality.fit_transform(data_clean['Nationality'])

# Defining X (features) and y (target)
X = data_clean.drop('Position', axis=1)  # Features
y = data_clean['Position']               # Target (classification target)

# Splitting into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Models: Logistic Regression, Decision Tree, Random Forest

# Logistic Regression Model
logreg_pipeline = Pipeline([
    ('scaler', StandardScaler()),             # Scaling numeric data
    ('clf', LogisticRegression(random_state=42))  # Logistic Regression Classifier
])

logreg_pipeline.fit(X_train, y_train)
logreg_y_pred = logreg_pipeline.predict(X_test)

# Decision Tree Model
dt_pipeline = Pipeline([
    ('scaler', StandardScaler()),             # Scaling numeric data
    ('clf', DecisionTreeClassifier(random_state=42))  # Decision Tree Classifier
])

dt_pipeline.fit(X_train, y_train)
dt_y_pred = dt_pipeline.predict(X_test)

# Random Forest Model
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),             # Scaling numeric data
    ('clf', RandomForestClassifier(random_state=42))  # Random Forest Classifier
])

rf_pipeline.fit(X_train, y_train)
rf_y_pred = rf_pipeline.predict(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define the hyperparameter grid for Logistic Regression
logreg_param_grid = {
    'clf__penalty': ['l1', 'l2', 'elasticnet', 'none'],  # Regularization penalty
    'clf__C': [0.01, 0.1, 1, 10],  # Regularization strength
    'clf__solver': ['liblinear', 'saga']  # Solver for optimization
}

# GridSearchCV for Logistic Regression
logreg_grid = GridSearchCV(logreg_pipeline, param_grid=logreg_param_grid, cv=5, n_jobs=-1, verbose=2)
logreg_grid.fit(X_train, y_train)

# Best hyperparameters
print("Best Hyperparameters for Logistic Regression:", logreg_grid.best_params_)

# Evaluate performance on test data
logreg_best_y_pred = logreg_grid.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, logreg_best_y_pred))
print("Classification Report:\n", classification_report(y_test, logreg_best_y_pred))
```

```python
# Define the hyperparameter grid for Decision Tree
dt_param_grid = {
    'clf__criterion': ['gini', 'entropy'],      # Function to measure split quality
    'clf__max_depth': [None, 10, 20, 30],       # Maximum depth of the tree
    'clf__min_samples_split': [2, 5, 10],       # Minimum number of samples to split a node
    'clf__min_samples_leaf': [1, 2, 4],         # Minimum number of samples required at a leaf node
}

# GridSearchCV for Decision Tree
dt_grid = GridSearchCV(dt_pipeline, param_grid=dt_param_grid, cv=5, n_jobs=-1, verbose=2)
dt_grid.fit(X_train, y_train)

# Best hyperparameters
print("Best Hyperparameters for Decision Tree:", dt_grid.best_params_)

# Evaluate performance on test data
dt_best_y_pred = dt_grid.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, dt_best_y_pred))
print("Classification Report:\n", classification_report(y_test, dt_best_y_pred))
```

```python
# Define the hyperparameter grid for Random Forest
rf_param_grid = {
    'clf__n_estimators': [100],           # Fixing number of trees
    'clf__max_depth': [None, 20],         # Fewer depth options
    'clf__min_samples_split': [2, 5],     # Fewer split options
    'clf__min_samples_leaf': [1, 2],      # Fewer leaf options
    'clf__bootstrap': [True]              # Fixing bootstrap to a single value
}


# GridSearchCV for Random Forest
rf_grid = GridSearchCV(rf_pipeline, param_grid=rf_param_grid, cv=5, n_jobs=1, verbose=2)
rf_grid.fit(X_train, y_train)


# Best hyperparameters
print("Best Hyperparameters for Random Forest:", rf_grid.best_params_)

# Evaluate performance on test data
rf_best_y_pred = rf_grid.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, rf_best_y_pred))
print("Classification Report:\n", classification_report(y_test, rf_best_y_pred))
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, auc
```

```python


# Function to evaluate and visualize model performance
def evaluate_model(y_test, y_pred, y_proba=None, classes_to_plot=None):
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    fig, ax = plt.subplots(figsize=(12, 10))  # Adjust the figure size for better readability
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues, ax=ax, values_format='.0f')
    plt.xticks(rotation=90, fontsize=8)  # Adjust font size if needed
    plt.yticks(fontsize=8)
    plt.title("Confusion Matrix")
    plt.show()

    # ROC-AUC Score and Curve (only for binary or multi-class probabilities)
    if y_proba is not None and len(np.unique(y_test)) <= 2:  # Binary classification check
        roc_auc = roc_auc_score(y_test, y_proba[:, 1])
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC)')
        plt.legend(loc="lower right")
        plt.show()

    # Multi-class ROC-AUC
    elif y_proba is not None and len(np.unique(y_test)) > 2:
        # Multi-class ROC-AUC Calculation
        roc_auc = roc_auc_score(y_test, y_proba, multi_class='ovr')
        print(f"ROC-AUC Score (Multiclass): {roc_auc:.2f}")

        # Plotting ROC for selected classes to avoid clutter
        if classes_to_plot is None:
            classes_to_plot = np.arange(min(3, y_proba.shape[1]))  # Default: plot first 3 classes
        plt.figure()
        for i in classes_to_plot:
            fpr, tpr, _ = roc_curve(y_test, y_proba[:, i], pos_label=i)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, lw=2, label=f'Class {i} ROC curve (area = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Multiclass ROC (Selected Classes)')
        plt.legend(loc="lower right")
        plt.show()

# Define the classes to plot in the ROC curve for multiclass (optional)
selected_classes = [0, 1, 2]  # Adjust or specify the classes you want to visualize
```

```python


# Logistic Regression
print("\nLogistic Regression Performance:")
logreg_best_y_pred = logreg_grid.predict(X_test)
logreg_best_y_proba = logreg_grid.predict_proba(X_test)
evaluate_model(y_test, logreg_best_y_pred, y_proba=logreg_best_y_proba, classes_to_plot=selected_classes)

# Decision Tree
print("\nDecision Tree Performance:")
dt_best_y_pred = dt_grid.predict(X_test)
dt_best_y_proba = dt_grid.predict_proba(X_test)
evaluate_model(y_test, dt_best_y_pred, y_proba=dt_best_y_proba, classes_to_plot=selected_classes)

# Random Forest
print("\nRandom Forest Performance:")
rf_best_y_pred = rf_grid.predict(X_test)
rf_best_y_proba = rf_grid.predict_proba(X_test)
evaluate_model(y_test, rf_best_y_pred, y_proba=rf_best_y_proba, classes_to_plot=selected_classes)
```
